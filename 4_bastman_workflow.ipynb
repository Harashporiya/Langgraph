```python
from langgraph.graph import StateGraph, START,END
from typing import TypedDict
```

```python
class BastmanState(TypedDict):

    runs:int
    balls:int
    fours:int
    sixes:int

    sr:float
    bpb:float
    boundary_percent:float
    summary:str
```

```python
def calculate_sr(state:BastmanState) -> BastmanState:

    sr = (state['runs']/state['balls'])*100

    return {'sr':sr}
```

```python
def calculate_bpb(state:BastmanState) -> BastmanState:

   bpb =  state['balls']/(state['fours']+state['sixes'])
   return {'bpb':bpb}
```

```python
def calculate_boundary_percent(state:BastmanState) -> BastmanState:

    boundary_percent = (((state['sixes']*4) + (state['fours']*6) ) / state['runs']) *100
    return {'boundary_percent':boundary_percent}
```

```python
def summary(state:BastmanState) -> BastmanState:

    summary=f"""Strike Rate - {state['sr']} \n
    Balls per boundary -  {state['bpb']} \n
    Boundary Percent - {state['boundary_percent']}
    """

    return {'summary':summary}
```

```python
graph = StateGraph(BastmanState)

graph.add_node('calculate_sr',calculate_sr)
graph.add_node('calculate_bpb',calculate_bpb)
graph.add_node('calculate_boundary_percent',calculate_boundary_percent)
graph.add_node('summary',summary)

graph.add_edge(START,'calculate_sr')
graph.add_edge(START,'calculate_bpb')
graph.add_edge(START,"calculate_boundary_percent")


graph.add_edge('calculate_sr', "summary")
graph.add_edge('calculate_bpb', "summary")
graph.add_edge("calculate_boundary_percent", "summary")

graph.add_edge("summary",END)


workflow= graph.compile()

```

```python
initial_state={
    'runs':100,
    'balls':50,
    'fours':6,
    'sixes':4
}

workflow.invoke(initial_state)
```
